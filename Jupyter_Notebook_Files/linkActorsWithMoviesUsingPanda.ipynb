```python
#importing libraries
from sqlalchemy import create_engine
import pandas as pd
```

```python
# creating connection string
engine = create_engine('postgresql+psycopg2://postgres:moviesondemand@moviesondemandaws.cfwjiare7kds.us-east-2.rds.amazonaws.com:5432/postgres')
```

```python
#creating a dataframe and loading data from AWS RDS
movie = pd.read_sql_table('Movies', con=engine)
actor = pd.read_sql_table('Names', con=engine)
link = pd.read_sql_table('title_principals', con=engine)
```

```python
#joining actor db with link table
actor_link = pd.merge(actor,link)
```

```python
#joining movie db with linked table
actor_movie_link = pd.merge(actor_link,movie)
```

```python
#visualization of resulting table
actor_movie_link.head()
```

```python
#sorting the dataframe
actor_movie_link.sort_values(by=['imdb_name_id'])
```
