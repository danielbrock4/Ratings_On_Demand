```python
import pandas as pd
import numpy as np
import plotly
import plotly.express as px
```

```python
resources_folder = r'../resources/'
ridge_test_results = resources_folder + 'ridge_test_results.csv'
```

```python
clean_comp_results = pd.read_csv(ridge_test_results, low_memory=False)
clean_comp_results.head()
```

```python

df = clean_comp_results
fig = px.scatter(df, x="Predicted Rating", y="Actual Rating", trendline="ols", hover_data=['Accuracy','Title', 'Genre' , 'Duration (min)' , 'Budget (mm)' , 'MPAA Rating' ])
fig.show()
```
