```python
# importing libraries to use
import pandas as pd
```

```python
# Create the path to IMDB movies
resources_folder = r'Resources/'
IMDb_file = resources_folder + 'IMDb_movies.csv'
# Reading the csv as a Panda dataframe
IMDb_movied_df = pd.read_csv(IMDb_file, low_memory=False)
IMDb_movied_df.head()
```

```python
# check the columns
IMDb_movied_df.columns.tolist()
```

```python
# keep only the ones we are going to use
IMDb_movied_df = IMDb_movied_df[['imdb_title_id','original_title','year','date_published','genre','duration','country','language','director','actors']]
```

```python
IMDb_movied_df.head()
```

```python
# change original_title to title
IMDb_movied_df.rename({'original_title' : 'title'}, axis='columns', inplace=True)
```

```python
IMDb_movied_df.head()
```

```python
# filter the df for country containing USA
df_country_filtered = IMDb_movied_df[IMDb_movied_df['country'].isin(['USA'])]
df_country_filtered.head()
```

```python
# filter the df for language containing english
df_language_filtered = df_country_filtered[df_country_filtered['language'].isin(['English'])]
df_language_filtered.head()
```

```python
# how many records do we have
df_language_filtered.count()
```

```python
# Check for null under an entire DataFrame due to directors and actors
df_language_filtered.isnull().values.any()
```

```python
# how many values are null
df_language_filtered.isnull().sum()
```

```python
# drop those null rows
not_null_df = df_language_filtered[df_language_filtered['director'].notnull()]
not_null_df = not_null_df[not_null_df['actors'].notnull()]
```

```python
# how many values are null and on which columns?
not_null_df.isnull().sum()
```

```python
# visualize
not_null_df.head()
```

```python
# how many records do we have
not_null_df.count()
```

```python
# do we have any NaN value?
not_null_df.isna().sum()
```

```python
# save final dataframe to a csv file
not_null_df.to_csv(resources_folder + 'clean_movies.csv', index=False)
```
